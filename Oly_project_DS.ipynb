```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("input/athlete_events.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
gender_count=df["Sex"].value_counts()
gender_count.plot(kind="pie",autopct='%.2f%%')
plt.title("M vs F ratio")
plt.show()                  
                  
```

```python
Medals_counts = df['Medal'].value_counts()
Medals_counts.plot(kind='pie',autopct="%.2f%%")
plt.title("% of medals won")
plt.show()
```

```python

df=pd.DataFrame(list(zip(df.Age.head(),df.Height.head(),df.Weight.head())),
               columns=["age","height","weight"],
               index=df.Year.head())
df.plot.bar()
plt.title("acc to year participation")
```

```python
df["Age"].plot.density()
plt.title("commom age")
```

```python
df['Year'].value_counts().head(10).plot(kind='bar')
plt.title('olympics')
plt.show()
```

```python
data=pd.DataFrame(df,columns=['Year','Age','Height','Weight'])
data.hist()
plt.show()
```

```python
df['City'].value_counts().head(10).plot(kind='bar')
plt.title('top ten city in olympic :}')
plt.show()
```

```python
df['Sport'].value_counts().head(20).plot(kind='bar')
plt.title('top 20 sports in olympics:}')
plt.show()
```

```python
df['Name'].value_counts().head(10).plot(kind='bar')
plt.title('top ten athlete in olympic :}')
plt.show()
```

```python
gd=df["Medal"]=="Gold"
gold=df[gd]
gold
```

```python
gold["Season"].value_counts().tail(20).plot.bar()
plt.title("most golds in which season?")
```

```python
gold["Name"].value_counts().head(20).plot.bar()
plt.title("people with most gold wins")
```

```python
sl=df["Medal"]=="Silver"
silver=df[sl]
silver
```

```python
silver["Name"].value_counts().head(20).plot.bar()
plt.title("people with most silver wins")
```

```python
bz=df["Medal"]=="Bronze"
bronze=df[bz]
bronze
```

```python
bronze["Name"].value_counts().head(20).plot.bar()
plt.title("people with most bronze wins")
```

```python
city_counts=df['City'].value_counts().head(10)
df=pd.DataFrame(list(zip(Medals_counts,city_counts)),
               columns=["medal","city"],
               index=df.Year.head())
df.plot.bar()
```

```python
df['Sport'].value_counts().head(20).plot(kind='bar')
plt.title('top 20 sports in olympics:}')
plt.show()
```

```python
df['Team'].value_counts().head(20).plot(kind='bar')
plt.title('top 20 teams in olympics')
plt.show()
```

```python
city_counts=df['City'].value_counts().tail(20)
df=pd.DataFrame(list(zip(Medals_counts.tail(20),city_counts(20))),
               columns=["medal","city"],
               index=df.Year.tail(20))
df.plot.bar()
```
